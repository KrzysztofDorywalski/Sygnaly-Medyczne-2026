### **Aspekty aplikacyjne FFT - aliasing**

$f_{alias} = |f-kf_s|$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import rfft, rfftfreq
```

```python
fs = 50   # ZA MAŁE próbkowanie
f_real = 40  # prawdziwa częstotliwość
```

```python
T = 1
N = fs * T

t = np.arange(N) / fs
signal = np.sin(2*np.pi*f_real*t)

# FFT
X = rfft(signal)
freqs = rfftfreq(N, 1/fs)

amp = (2.0 / N) * np.abs(X)
```

```python
plt.stem(freqs, amp)
plt.title("Aliasing w FFT")
plt.xlabel("Hz")
plt.grid()
plt.show()
```
