### **Aspekty aplikacyjne FFT - rozdzielczość czestotliwościowa**

Rozdzielczość częstotliwościowa FFT:

$\Delta f = \frac{f_s}{N}$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import rfft, rfftfreq
```

```python
fs = 100  # Hz
```

```python
def compute_fft(N):
    t = np.arange(N) / fs
    
    signal = (
        np.sin(2*np.pi*10*t) +
        np.sin(2*np.pi*12*t)
    )
    
    X = rfft(signal)
    freqs = rfftfreq(N, 1/fs)
    
    amplitude = (2.0 / N) * np.abs(X)
    amplitude[0] /= 2
    
    return freqs, amplitude
```

```python
# różne długości
Ns = [100, 200, 1000]
```

```python
plt.figure(figsize=(12,8))

for i, N in enumerate(Ns):
    freqs, amp = compute_fft(N)
    
    plt.subplot(len(Ns), 1, i+1)
    plt.stem(freqs, amp)
    plt.xlim(0, 30)
    plt.title(f"N = {N}, Δf = {fs/N:.2f} Hz")
    plt.grid()

plt.tight_layout()
plt.show()
```
