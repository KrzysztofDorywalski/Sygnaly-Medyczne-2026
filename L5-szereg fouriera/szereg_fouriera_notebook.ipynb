# Szereg Fouriera – materiał dydaktyczny
### Przykład: rozwinięcie funkcji f(t)=t w przedziale (-π, π)

Ten notebook pokazuje krok po kroku:
- ideę szeregu Fouriera
- obliczenie współczynników
- wizualizację przybliżeń funkcji

Funkcja:

$$f(t)=t, \quad -\pi < t < \pi$$

Przedłużamy ją okresowo z okresem $2\pi$.

## Import bibliotek

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Definicja funkcji oryginalnej

```python
def f(t):
    return t
```

## Współczynniki szeregu Fouriera

Dla tej funkcji:

$$b_n = \frac{2(-1)^{n+1}}{n}$$

Szereg:

$$f(t)=2\sum_{n=1}^{\infty}\frac{(-1)^{n+1}}{n}\sin(nt)$$

```python
def fourier_series(t, N):
    s = 0
    for n in range(1, N+1):
        bn = 2*((-1)**(n+1))/n
        s += bn*np.sin(n*t)
    return s
```

## Wykres funkcji oryginalnej

```python
t = np.linspace(-np.pi, np.pi, 1000)
y = f(t)

plt.figure()
plt.plot(t, y)
plt.title("Funkcja f(t)=t")
plt.xlabel("t")
plt.ylabel("f(t)")
plt.show()
```

## Przybliżenie szeregiem Fouriera (N=5)

```python
approx = fourier_series(t, 5)

plt.figure()
plt.plot(t, y)
plt.plot(t, approx)
plt.title("Przybliżenie szeregiem Fouriera N=5")
plt.xlabel("t")
plt.ylabel("wartość")
plt.show()
```

## Przybliżenie szeregiem Fouriera (N=20)

```python
approx = fourier_series(t, 20)

plt.figure()
plt.plot(t, y)
plt.plot(t, approx)
plt.title("Przybliżenie szeregiem Fouriera N=20")
plt.xlabel("t")
plt.ylabel("wartość")
plt.show()
```
