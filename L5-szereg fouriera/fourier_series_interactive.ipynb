# Interaktywny szereg Fouriera
Użyj suwaka, aby zmieniać liczbę harmonicznych i obserwować przybliżenie.

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, IntSlider
```

```python
def f(t):
    return t

def fourier_series(t, N):
    s = 0
    for n in range(1, N+1):
        bn = 2*((-1)**(n+1))/n
        s += bn*np.sin(n*t)
    return s
```

```python
t = np.linspace(-np.pi, np.pi, 1000)

def plot_fourier(N):
    plt.figure()
    plt.plot(t, f(t))
    plt.plot(t, fourier_series(t, N))
    plt.title(f"Szereg Fouriera, N={N}")
    plt.show()

interact(plot_fourier, N=IntSlider(min=1, max=50, step=1, value=5));
```
