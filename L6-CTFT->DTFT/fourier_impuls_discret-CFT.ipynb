# Od szeregu Fouriera do transformaty – impuls prostokątny
Notebook pokazuje przejście od widma dyskretnego do ciągłego.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = 1
tau = 1

def rect(t):
    return np.where(np.abs(t) <= tau/2, A, 0)
```

```python
def c_n(n, T):
    w0 = 2*np.pi / T
    if n == 0:
        return A * tau / T
    else:
        return (A/T) * (2*np.sin(n*w0*tau/2)/(n*w0))
```

```python
T = 10
N = 20

n_vals = np.arange(-N, N+1)
c_vals = np.array([c_n(n, T) for n in n_vals])
freqs = n_vals * (2*np.pi / T)

plt.figure()
plt.stem(freqs, np.abs(c_vals))
plt.title('Widmo dyskretne')
plt.show()
```

```python
omega = np.linspace(-20, 20, 1000)
F = A*tau * np.sinc(omega*tau/(2*np.pi))

plt.figure()
plt.plot(omega, np.abs(F))
plt.title('Widmo ciągłe')
plt.show()
```
